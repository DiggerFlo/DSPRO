# Import Pandas

```python
import pandas as pd
```

```python
#Because there are ; in the files aswell we have to escape
def load_dataset(path):
    rows = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            parts = line.split(";", 1)
            if len(parts) == 2:
                rows.append({"category": parts[0], "text": parts[1]})
    return pd.DataFrame(rows)

train = load_dataset("../data/raw/train.csv")
articles  = load_dataset("../data/raw/articles.csv")
test = load_dataset("../data/raw/test.csv")

print(f"train: {len(train)} Artikel")
print(f"articles:  {len(articles)} Artikel")
print(f"test: {len(test)} Artikel")
print()
print("Kategorien:")
print(train["category"].value_counts())
print()
print("Beispiel:")
print(train.iloc[0]["category"])
print(train.iloc[0]["text"][:300])
```
